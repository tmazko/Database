```python
import duckdb
import pandas as pd
import altair as alt

```

```python

df = pd.read_csv('years_stat.csv', names=['year', 'count', 'avg_price'])
years = list(range(2006, 2026, 2))
```

```python
chart = alt.Chart(df).mark_line(
    color='#6A0DAD',
    strokeWidth=3,
    point={
        'color': 'orange',
        'fill': 'orange',
        'size': 70
    }
).encode(
    x=alt.X('year:O',
            axis=alt.Axis(
                values=years,
                title='Release Year',
                titleFontSize=16,
                labelFontSize=14,
                labelAngle=0
            )
    ),
    y=alt.Y('count',
            axis=alt.Axis(
                title='Number of Games Released',
                titleFontSize=16,
                labelFontSize=14
            )
    ),
    tooltip=[
        alt.Tooltip('year', title='Year'),
        alt.Tooltip('count', title='Game Count'),
        alt.Tooltip('avg_price', title='Avg Price', format='$.2f')
    ]
).properties(
    title=alt.TitleParams(
        text='Dynamics of Game Releases & Average Price',
        fontSize=24,
        subtitle='Source: Steam Dataset Analysis',
        subtitleFontSize=14
    ),
    width=800,
    height=450
).interactive()

chart.save('year_statistics_chart.html')
chart
```
